### DBSCAN

```python
from importlib import reload
# Auto-reload
import ml_course.dbscan
DBSCAN = reload(ml_course.dbscan).DBSCAN

model = DBSCAN(eps=1, min_samples=5)
y_pred = model.fit_predict(X)

print(model.labels_)
print(model.status_)

from matplotlib.pyplot import scatter

%matplotlib inline
scatter(X[:, 0], X[:, 1], c=y_pred, cmap='rainbow')
```

`CORE_POINT`, `BORDER_POINT` atau `OUTLIER`?

```python
from matplotlib.pyplot import legend, scatter

status_info = (
    (DBSCAN.OUTLIER, 'outlier', 'red'),
    (DBSCAN.CORE_POINT, 'core point', 'cyan'),
    (DBSCAN.BORDER_POINT, 'border point', 'blue')
)

%matplotlib inline
for i, label, color in status_info:
    X_status = X[model.status_ == i]
    scatter(X_status[:, 0], X_status[:, 1], c=color, label=label)
legend()
```
